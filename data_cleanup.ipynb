```python
import pandas as pd
```

```python
#read data from csv
data_01 = pd.read_csv("citibike_downloads/201901-citibike-tripdata.csv")
data_02 = pd.read_csv("citibike_downloads/201902-citibike-tripdata.csv")
data_03 = pd.read_csv("citibike_downloads/201903-citibike-tripdata.csv")
data_04 = pd.read_csv("citibike_downloads/201904-citibike-tripdata.csv")
data_05 = pd.read_csv("citibike_downloads/201905-citibike-tripdata.csv")
data_06 = pd.read_csv("citibike_downloads/201906-citibike-tripdata.csv")
data_07 = pd.read_csv("citibike_downloads/201907-citibike-tripdata.csv")
data_08 = pd.read_csv("citibike_downloads/201908-citibike-tripdata.csv")
data_09 = pd.read_csv("citibike_downloads/201909-citibike-tripdata.csv")
data_10 = pd.read_csv("citibike_downloads/201910-citibike-tripdata.csv")
data_11 = pd.read_csv("citibike_downloads/201911-citibike-tripdata.csv")
data_12 = pd.read_csv("citibike_downloads/201912-citibike-tripdata.csv")
```

```python
data_combined = pd.concat([data_01, data_02, data_03, data_04, data_05, data_06, data_07, data_08, data_09,
                           data_10, data_11, data_12], ignore_index=True)
```

```python
data_combined2 = data_combined.copy()
data_combined2
```

```python
# split datetime field into multiple to get dates and times separate
data_combined2[['date','time']] = data_combined2["starttime"].str.split(" ", 1, expand=True)
```

```python
data_combined2.head(5)
```

```python
# get rid of minutes and seconds - keep hours (military time)
data_combined2['time'] = data_combined2["time"].str.split(":", 1, expand=True)[0]
```

```python
data_combined2.head(5)
```

```python
# drop unwanted columns
data_combined2.drop(columns =['starttime', 'stoptime', 'start station id','end station id',
                          'end station latitude','end station longitude', 'bikeid'], inplace=True)
```

```python
data_combined2.head()
```

```python
# rename columns to assist with header flattening later 
data_combined2.rename(columns={'time': 'hour_of_day', 'start station name': 'start_sta_name',
                             'start station latitude': 'start_sta_lat', 
                             'start station longitude': 'start_sta_lon',
                             'end station name': 'end_sta_name', 'birth year': 'birth_year'},
                             inplace=True)
```

```python
data_combined2.rename(columns={'start station longitude': 'start_sta_lon',
                            'birth year': 'birth_year'},
                             inplace=True)
```

```python
data_combined2
```

```python
# aggregate trip duration data based on hour of day, date and gender
gender_trips = data_combined2.groupby(['hour_of_day', 'date', 'gender']).agg(
    {'tripduration':['min', 'max', 'mean', 'sum', 'count']})
```

```python
gender_trips.head(5)
```

```python
# aggregate trip duration data based on hour of day, date and usertype
usertype_trips = data_combined2.groupby(['hour_of_day', 'date', 'usertype']).agg(
    {'tripduration':['min', 'max', 'mean', 'sum', 'count']})
```

```python
usertype_trips.head(5)
```

```python
# aggregate trip duration data based on hour of day, date and birthyear
age_trips = data_combined2.groupby(['hour_of_day', 'date', 'birth_year']).agg(
    {'tripduration':['min', 'max', 'mean', 'sum', 'count']})
```

```python
age_trips.head(5)
```

```python
# aggregate number of trips vs date
rides_day = data_combined2.groupby(['date']).agg(
    {'tripduration':['count']})
```

```python
rides_day.head()
```

```python
#  separate agg for start station name
start_sta = data_combined2.groupby(['start_sta_name', 'usertype']).agg({'tripduration':['min', 'max', 'mean', 'sum', 'count'],
                                            'birth_year':['min', 'max'],
                                            'start_sta_lat': 'min',
                                            'start_sta_lon': 'min'})
```

```python
start_sta.head()
```

```python
#  separate agg for start station name by birth year
start_sta_age = data_combined2.groupby(['start_sta_name', 'birth_year']).agg({'tripduration':['min', 'max', 'mean', 'sum', 'count'],
                                            'start_sta_lat': 'min',
                                            'start_sta_lon': 'min'})
```

```python
start_sta_age.head()
```

```python
usertype_trips.columns = [f"{x}_{y}" for x, y in usertype_trips.columns.to_flat_index()]
usertype_trips.to_csv("Resources/usertype_trips.csv", index=True, header=True)
```

```python
age_trips.columns = [f"{x}_{y}" for x, y in age_trips.columns.to_flat_index()]
age_trips.to_csv("Resources/age_trips.csv", index=True, header=True)
```

```python
rides_day.columns = [f"{x}_{y}" for x, y in rides_day.columns.to_flat_index()]
rides_day.to_csv("Resources/rides_day.csv", index=True, header=True)
```

```python
start_sta.columns = [f"{x}_{y}" for x, y in start_sta.columns.to_flat_index()]
start_sta.to_csv("Resources/start_sta.csv", index=True, header=True)
```
